```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
folders = os.listdir()
```

```python
periods = os.listdir('PETR3')
```

```python
folders[2:]
```

```python
words = ['PETR3', 'PETR4', 'Petrobras', 'Petróleo']
```

```python
dataframes = dict()
```

```python
pd.DataFrame()
```

```python
for w in words:
    dataframes[f] = pd.DataFrame()
    for p in periods:
        filepath = '''{}/{}'''.format(w, p)
        dataframes[f] = pd.concat([dataframes[f], pd.read_csv(filepath, skiprows=1)])
```

```python
df = dataframes['PETR3'].merge(dataframes['PETR4'], how='left', on='Day').merge(
    dataframes['Petrobras'], how='left', on='Day').merge(dataframes['Petróleo'], how='left', on='Day')
```

```python
df
```

```python
df.plot(x="Day", y = df.columns[1:], kind="line", figsize=(10,5))
# plt.figure(figsize=(100, 4)) 
plt.show();
```
